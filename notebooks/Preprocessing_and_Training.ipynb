```python
#Import prrequesites
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# Assign filename: file
#project location on local
Project_Local ='C:/Users/Greki/Capstone_Two/'
File_Location='data/interim/targeted_counties_chickenpox_data.csv'
file =Project_Local+File_Location
# Load chickenpox data
df1 = pd.read_csv(file)
#datatype for Date column to date time
df1['Date'] = pd.to_datetime(df1['Date'], format='%Y/%m/%d')
```

```python
df1.info()
```

```python
#delete columns with county names, old index, and year
df2= df1.drop(['County_C0','County_C1', 'County_C2', 'County_C3','Unnamed: 0','Year'], axis =1)
#sent a new time base index
df2.index = df2['Date']
#datatype for Date column to date time
df2.sort_index(inplace = True)
df3= df2.drop(['Date'],axis=1 )
df3.head()
#Set variables and responses
X = df3.drop(labels=['Cases_C0'], axis=1)
y = df3['Cases_C0']
X.head(5)
```

```python
#Create TimeSeriesSplit
from sklearn.model_selection import TimeSeriesSplit
TSS= TimeSeriesSplit(n_splits = 4)
for train_index, test_index in TSS.split(X):
    X_train, X_test = X.iloc[train_index, :], X.iloc[test_index,:]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
```

```python
#Import Standard Scaler
from sklearn.preprocessing import StandardScaler
```

```python
#Set scaler for X_train
scaler = StandardScaler().fit(X_train)
```

```python
scaler.mean_
```

```python
#Transform X_train scaled data
X_train_scaled=scaler.transform(X_train)
```

```python
#X_train scaled mean
print(X_train_scaled.mean(axis=0))
```

```python
#X_trained Scaled Standard deviation
print(X_train_scaled.std(axis=0))
```

```python
#Set scaler for X_test
scaler_X_test = scaler.fit(X_test)
```

```python
#Transform X_test scaled data
X_test_scaled=scaler_X_test.transform(X_test)
```

```python
#X_test scaled mean
print(X_test_scaled.mean (axis=0))
```

```python
#X_test Scaled Standard deviation
print(X_test_scaled.std(axis=0))
```
